```python
from google.colab import files
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# ðŸ”¹ Step 1: Upload image
uploaded = files.upload()

# ðŸ”¹ Step 2: Auto-detect uploaded filename
filename = next(iter(uploaded))
print("âœ… Uploaded file:", filename)

# ðŸ”¹ Step 3: Read the image
image = cv2.imread(filename)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# ðŸ”¹ Step 4: Create a structuring element (5x5 rectangular)
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))

# ðŸ”¹ Step 5: Perform morphological opening and closing
opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)

# ðŸ”¹ Step 6: Convert results for display
opening_rgb = cv2.cvtColor(opening, cv2.COLOR_BGR2RGB)
closing_rgb = cv2.cvtColor(closing, cv2.COLOR_BGR2RGB)

# ðŸ”¹ Step 7: Display all images
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(opening_rgb)
plt.title("Opening Operation")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(closing_rgb)
plt.title("Closing Operation")
plt.axis("off")

plt.show()
```
